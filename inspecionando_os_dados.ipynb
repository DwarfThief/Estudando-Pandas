# Estudando sobre a biblioteca Pandas

### Início:

* Começamos importando a biblioteca
    * é comum apelidar a lib como `pd`
* Definimos o DataFrame lendo o arquivo desejado

```python
import pandas as pd

df = pd.read_csv('./Sample-Data/intel.csv')
```

### Imprimindo o DataFrame para ter certeza que o Pandas abriu o documento

```python
print(df)
```

### Imprime as colunas do DataFrame

```python
print(df.columns)
```

### Imprime das linhas 0 até as que foram passadas pelo parâmetro, nesse caso, 10

```python
print(df.head(10))
```

### Imprime das linhas finais para cima

```python
print(df.tail(10))
```

### Imprime as informações do documento, como os valores das colunas, etc

```python
print(df.info())
```

### Selecionar toda a coluna passada por parâmetro o seu nome

```python
open = df['Open']
print(open)
```

### Imprime as duas ou mais colunas com os nomes passados no parâmetro

```python
print(df[['Open', 'Close']])
```

### Imprime a descrição do arquivo, com valor máximo, mínimo e as médias de valores, tudo por coluna

```python
print(df.describe())
```

### Checando quais valores da coluna cumprem a condição

```python
print(df['Open'] > 50)
```

### Retorna apenas aqueles que cumprem a condição

```python
print(df[df['Open']>50])
```
