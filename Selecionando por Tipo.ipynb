# Selecionando atráves do tipo

```python
import pandas as pd
df = pd.read_csv('./Sample-Data/outrosFormatos/aluguel.csv', sep=';')
df.head(10)
```

```python
list(df['Tipo'].drop_duplicates())
```

```python
# Vamos classificar os imóveis relacionados
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
selecao = df['Tipo'].isin(casa)
selecao
```

```python
df_casa = df[selecao]
df_casa
```

```python
df.shape[0]
```

```python
df_casa.shape[0]
```

```python
df_casa.index = range(df_casa.shape[0])
df_casa
```

## Exportando o DataFrame

```python
df_casa.to_csv('./Sample-Data/DataFrameCasa.csv', sep=';', index=False)
```
